```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("https://archive.ics.uci.edu/static/public/14/data.csv")
```

```python
df = pd.DataFrame(data)
df
```

```python
#melt
columns_to_melt = ['menopause','node-caps','breast','breast-quad','irradiat','Class']


df_melted = df.melt(id_vars=columns_to_melt, var_name='Measurement', value_name='Value')
df_melted
```

```python
#pivot
df_pivoted = df_melted.pivot_table(index=columns_to_melt, columns=['Measurement'], values=['Value'],aggfunc='first')
df_pivoted.reset_index(inplace=True)

df_pivoted.columns = ['age', 'menopause', 'tumor-size', 'breast', 'Class', 'inv-nodes', 'node-caps', 'deg-malig', 'breast-quad', 'irradiat']


df_pivoted
```

```python
#aggregation
df['deg-malig'].mean()
```

```python
#iterate
for index, row in df_pivoted.iterrows():
    print(index, row['age'])
```

```python
#groupby
df.groupby('Class').agg({
    'deg-malig': ['min', 'max','mean','std']
})
```
