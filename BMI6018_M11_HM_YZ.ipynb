```python
#  Altair's development is led by the Interactive Data Lab at the University of Washington. It is an open-source project.
# source of altair: https://altair-viz.github.io/index.html
# introduction to altair: On this official website of altair, they introduce altair as: 
    # "Vega-Altair is a declarative visualization library for Python. Its simple, friendly and consistent API, built on top of the powerful Vega-Lite grammar, empowers you to spend less time writing code and more time exploring your data."
    # it can create different kinds of figures or plots based on the data for visualization through the determination of visual encoding channels. Other complicated inspects of the plots can be managed automatically. 
```

```python
# Advantages: 
    # Altair is a relatively straightforward package for data visualization. People can use it to generate the desired plots without the need for complex codes.
    # Altair can automatically interprets data types and generate plots for visualization needs, which reduce the manual work load.
    # Altair integrates with Panda dataframes, which simplify the process of loading, manipulating, and visualize the data.
```

```python
# Disadvantages: 
    # The types of plot that Altair can generate is limited. Some advanced format of data visualization cannot be generated using Altair.
    # There is limited space for the customization of the plots generated by Altair, which makes it hard when people want to change some details of the plots or chart.
    # Altair is a good choice for the simple data visualization, but it is hard to generate complex visualization using Altair. 
```

```python
import altair as alt
import pandas as pd
```

```python
weather_data_df = pd.read_csv('weather.csv')
df = weather_data_df.fillna(0)
```

```python
import altair as alt
import pandas as pd

scatter_plot = alt.Chart(df).mark_point().encode(
    x='temp:Q',
    y='humid:Q',
)

alt.renderers.enable('html')  


scatter_plot.save('scatter_plot.html')

```
